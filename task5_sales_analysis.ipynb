```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    'Date': ['2025-01-01', '2025-01-01', '2025-01-02', '2025-01-02', '2025-01-03'],
    'Product': ['A', 'B', 'A', 'C', 'B'],
    'Quantity': [5, 3, 2, 4, 6],
    'Price': [100, 200, 100, 150, 200]
}
df = pd.DataFrame(data)


print("First 5 rows:")
print(df.head())

df['Total'] = df['Quantity'] * df['Price']
sales_per_product = df.groupby('Product')['Total'].sum()
print("\nSales per product:")
print(sales_per_product)

sales_per_product.plot(kind='bar', title="Sales per Product", color='skyblue')
plt.ylabel("Total Sales ()")
plt.show()

daily_sales = df.groupby('Date')['Total'].sum()
print("\nDaily Sales:")
print(daily_sales)

daily_sales.plot(kind='line', marker='o', title="Daily Sales")
plt.ylabel("Total Sales ()")
plt.show()
```
